# EDA Dataset: Enhanced Pizza Sales 2024-2025
## Análisis de criterios de evaluación

Este notebook recoge el análisis de los criterios de evaluación del 
dataset Enhanced Pizza Sales 2024-2025 como parte del proceso de 
selección del dataset para el TFG de predicción de demanda en 
pizzerías con delivery.

```python
import pandas as pd

# Carga del dataset
df = pd.read_excel('../../data/0_seleccion_dataset/kaggle_enhanced_pizza/Enhanced_pizza_sell_data_2024-25.xlsx')

print(f"Dataset cargado correctamente")

print("--- CRITERIO 1: VOLUMEN")
print(f"Registros totales: {df.shape[0]}")

# Pedidos por día
df['Order Time'] = pd.to_datetime(df['Order Time'])
df['date'] = df['Order Time'].dt.date
pedidos_dia = df.groupby('date').size()
print(f"Media de pedidos por día: {pedidos_dia.mean():.1f}")
print(f"Máximo pedidos en un día: {pedidos_dia.max()}")
print(f"Mínimo pedidos en un día: {pedidos_dia.min()}")
```

## Criterio 1: Volumen
El dataset solo contiene 1.004 registros con una media de 1,2 pedidos 
por día, volumen completamente insuficiente para entrenar un modelo 
de series temporales.

```python
print("--- CRITERIO 2: GRANULARIDAD TEMPORAL")
print(f"Columnas disponibles: {df.columns.tolist()}")
print(f"¿Tiene fecha? {'Order Time' in df.columns}")
print(f"¿Tiene hora? {'Order Time' in df.columns}")
print(f"\nEjemplo de valor: {df['Order Time'].iloc[0]}")
```

## Criterio 2: Granularidad temporal
El dataset dispone de fecha y hora exacta en la columna `Order Time`.

```python
print("--- CRITERIO 3: RANGO TEMPORAL")
print(f"Fecha mínima: {df['Order Time'].min()}")
print(f"Fecha máxima: {df['Order Time'].max()}")
dias = (df['Order Time'].max() - df['Order Time'].min()).days
print(f"Días cubiertos: {dias}")
print(f"Años cubiertos: {dias/365:.1f}")
```

## Criterio 3: Rango temporal
El dataset cubre aproximadamente 2,5 años (2024-2026), pero con 
solo 1.004 registros repartidos en ese periodo la densidad de datos 
es demasiado baja para construir una serie temporal fiable.

```python
print("--- CRITERIO 4: CALIDAD DE DATOS")
print(f"Valores nulos:\n{df.isnull().sum()}")
print(f"\nDuplicados: {df.duplicated().sum()}")
```

## Criterio 4: Calidad de los datos
El dataset no contiene valores nulos ni registros duplicados.

```python
print("--- CRITERIO 5: ESPECIFICIDAD DE RESTAURACIÓN ")
print(f"Restaurantes distintos: {df['Restaurant Name'].nunique()}")
print(f"\nRestaurantes:")
print(df['Restaurant Name'].value_counts())
print(f"\nCiudades distintas: {df['Location'].nunique()}")
print(f"\nCiudades:")
print(df['Location'].value_counts().head(10))
```

## Criterio 5: Especificidad de restauración
Los datos son de pizzerías pero de diferentes restaurantes y de diferentes zonas de EEUU, lo que dificulta el entrenamiento del modelo.

## Verificación de criterios de evaluación

- ❌ **Volumen:** solo 1.004 registros con una media 
  de 1,2 pedidos por día resulta insuficiente para entrenar el modelo.
- ✅ **Fecha y hora:** dispone de fecha y hora exacta.
- ⚠️ **Rango temporal:** 2,5 años pero con densidad de datos 
  insuficiente.
- ✅ **Calidad de datos:** sin valores nulos ni duplicados.
- ❌ **Especificidad:** 6 cadenas distintas en 84 ciudades, no 
  representa un único establecimiento.

## Conclusión: dataset descartado

El análisis confirma que este dataset no cumple los criterios 
mínimos de selección.

El criterio de volumen y el de especificidad son determinantes 
para el descarte, ya que hacen inviable construir una serie 
temporal representativa de un único establecimiento.

**Decisión: dataset descartado.**
