# EDA Dataset: Store Item Demand Forecasting
## Análisis de criterios de evaluación

Este notebook recoge el análisis de los criterios de evaluación del 
dataset Store Item Demand Forecasting como parte del proceso de 
selección del dataset para el TFG de predicción de demanda en 
pizzerías con delivery.

```python
import pandas as pd

# Carga del dataset
df = pd.read_csv('../data/kaggle_store_item_demand/train.csv')

print(f"Dataset cargado correctamente")

print("=== CRITERIO 1: VOLUMEN ===")
print(f"Registros totales: {df.shape[0]}")
print(f"Tiendas distintas: {df['store'].nunique()}")
print(f"Productos distintos: {df['item'].nunique()}")
print(f"Registros por tienda y producto: {df.shape[0] / (df['store'].nunique() * df['item'].nunique()):.0f}")
```

## Criterio 1: Volumen
⚠️ El dataset tiene 913.000 registros en total, pero están repartidos 
entre 10 tiendas y 50 productos distintos. Por tienda y producto 
la serie temporal tiene 1.826 registros, equivalente a 5 años de 
datos diarios.

```python
print("=== CRITERIO 2: GRANULARIDAD TEMPORAL ===")
print(f"Columnas disponibles: {df.columns.tolist()}")
print(f"¿Tiene fecha? {'date' in df.columns}")
print(f"¿Tiene hora? {'time' in df.columns or 'hour' in df.columns}")
print(f"\nEjemplo de valor de fecha: {df['date'].iloc[0]}")
```

## Criterio 2: Granularidad temporal
❌ El dataset solo dispone de fecha diaria, sin información sobre 
la hora del pedido. Esto impide analizar los patrones de demanda 
por franja horaria, variable fundamental identificada en el 
estado del arte.

```python
print("=== CRITERIO 3: RANGO TEMPORAL ===")
df['date'] = pd.to_datetime(df['date'])
print(f"Fecha mínima: {df['date'].min()}")
print(f"Fecha máxima: {df['date'].max()}")
dias = (df['date'].max() - df['date'].min()).days
print(f"Días cubiertos: {dias}")
print(f"Años cubiertos: {dias/365:.1f}")
```

## Criterio 3: Rango temporal
✅ El dataset cubre 5 años completos (2013-2017), un rango temporal 
amplio y suficiente para entrenar un modelo de series temporales.

```python
print("=== CRITERIO 4: CALIDAD DE DATOS ===")
print(f"Valores nulos:\n{df.isnull().sum()}")
print(f"\nDuplicados: {df.duplicated().sum()}")
```

## Criterio 4: Calidad de los datos
✅ El dataset no presenta valores nulos ni registros duplicados.

```python
print("=== CRITERIO 5: ESPECIFICIDAD DE RESTAURACIÓN ===")
print(f"¿Es específico de restauración? No")
print(f"\nColumnas disponibles: {df.columns.tolist()}")
print(f"\nEjemplo de productos (item):")
print(df['item'].unique()[:10])
print(f"\nEjemplo de tiendas (store):")
print(df['store'].unique())
```

## Criterio 5: Especificidad de restauración
❌ El dataset no es específico de restauración. Corresponde a ventas 
de productos genéricos en 10 tiendas minoristas distintas, sin 
ninguna relación con el sector de la restauración o el delivery.

## Verificación de criterios de evaluación

- ⚠️ **Volumen:** 913.000 registros totales pero repartidos entre 
  10 tiendas y 50 productos distintos.
- ❌ **Granularidad temporal:** solo fecha diaria, sin hora del pedido.
- ✅ **Rango temporal:** 5 años completos (2013-2017).
- ✅ **Calidad de datos:** sin valores nulos ni duplicados.
- ❌ **Especificidad:** productos genéricos en tiendas minoristas, 
  sin relación con restauración o delivery.

## Conclusión: dataset descartado

El análisis confirma que este dataset no cumple los criterios 
mínimos de selección:

- ⚠️ **Volumen:** amplio en total pero fragmentado entre múltiples 
  tiendas y productos.
- ❌ **Granularidad temporal:** ausencia de hora del pedido, variable 
  fundamental para el análisis de franjas horarias.
- ✅ **Rango temporal:** 5 años completos.
- ✅ **Calidad de datos:** sin nulos ni duplicados.
- ❌ **Especificidad:** no tiene relación con el sector de la 
  restauración ni el delivery.

Los criterios de granularidad temporal y especificidad son 
determinantes para el descarte, ya que la ausencia de hora 
impide analizar los patrones de demanda por franja horaria 
y el contexto no es representativo del negocio estudiado.

**Decisión: dataset descartado.**
