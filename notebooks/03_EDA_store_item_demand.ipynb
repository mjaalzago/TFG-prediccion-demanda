# EDA Dataset: Store Item Demand Forecasting
## Análisis previo al descarte

Este notebook recoge el análisis exploratorio del dataset Store Item 
Demand Forecasting como parte del proceso de selección del dataset para 
el TFG de predicción de demanda en pizzerías con delivery.

```python
import pandas as pd

# Carga del dataset completo
df = pd.read_csv('../data/kaggle_store_item_demand/train.csv')

print(f"Registros: {df.shape[0]}")
print(f"Columnas: {df.shape[1]}")
print(f"\nColumnas disponibles:")
print(df.columns.tolist())
print(f"\nPrimeras filas:")
print(df.head())
```

```python
# Análisis de la variable temporal
print(f"Rango temporal: {df['date'].min()} → {df['date'].max()}")
print(f"\nNúmero de tiendas distintas: {df['store'].nunique()}")
print(f"Número de productos distintos: {df['item'].nunique()}")
print(f"\n¿Tiene columna de hora? {'time' in df.columns or 'hour' in df.columns}")
```

## Conclusión: descarte del dataset

El análisis exploratorio confirma que este dataset no es adecuado 
para el objetivo del proyecto por los siguientes motivos:

- **No es específico de restauración:** los datos corresponden a 
  ventas de productos genéricos en 10 tiendas minoristas distintas, 
  sin ninguna relación con el sector de la restauración o el delivery.
- **Sin granularidad horaria:** solo dispone de fecha diaria, sin 
  información sobre la hora del pedido. Esto impide analizar los 
  patrones de demanda por franja horaria, variable fundamental 
  identificada en el estado del arte.
- **Datos heterogéneos:** mezcla 50 productos distintos en 10 tiendas 
  diferentes, lo que no representa el comportamiento de un único 
  establecimiento.
- **Rango temporal:** aunque cubre 5 años (2013-2017), la ausencia 
  de la hora en la que se produce la venta y el contexto lo hacen 
  inviable para este proyecto.

**Decisión: dataset descartado.**
