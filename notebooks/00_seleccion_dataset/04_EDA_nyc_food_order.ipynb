# EDA Dataset: NYC Restaurant Food Order & Delivery
## Análisis de criterios de selección

Este notebook recoge el análisis exploratorio del dataset NYC Restaurant 
Food Order & Delivery como parte del proceso de selección del dataset 
para el TFG de predicción de demanda en pizzerías con delivery.

```python
import pandas as pd

# Carga del dataset
df = pd.read_csv('../../data/0_seleccion_dataset/kaggle_NYC_restaurants_data/food_order.csv')

print(f"Dataset cargado correctamente")

print("--- CRITERIO 1: VOLUMEN ")
print(f"Registros totales: {df.shape[0]}")
print(f"\nRestaurantes distintos: {df['restaurant_name'].nunique()}")
print(f"Tipos de cocina distintos: {df['cuisine_type'].nunique()}")
```

## Criterio 1: Volumen
El dataset solo contiene 1.898 registros siendo insuficiente
para entrenar el modelo.

```python
print("--- CRITERIO 2: GRANULARIDAD TEMPORAL ")
print(f"Columnas disponibles: {df.columns.tolist()}")
print(f"¿Tiene fecha? {'date' in df.columns}")
print(f"¿Tiene hora? {'time' in df.columns or 'hour' in df.columns}")
print(f"\nVariable temporal disponible: day_of_the_week")
print(f"Valores únicos: {df['day_of_the_week'].unique()}")
```

## Criterio 2: Granularidad temporal
El dataset no dispone de fecha ni hora de los pedidos. La única 
información temporal es si el pedido se realizó en fin de semana 
o entre semana, con solo dos valores posibles: Weekend y Weekday. 
Esto hace imposible construir una serie temporal.

```python
print("--- CRITERIO 3: RANGO TEMPORAL ")
print(f"¿Tiene columna de fecha? {'date' in df.columns}")
print(f"\nDistribución por tipo de día:")
print(df['day_of_the_week'].value_counts())
```

## Criterio 3: Rango temporal
Al no disponer de fecha real no es posible determinar el rango 
temporal del dataset ni construir una serie temporal ordenada 
en el tiempo.

```python
print("--- CRITERIO 4: CALIDAD DE DATOS ")
print(f"Valores nulos:\n{df.isnull().sum()}")
print(f"\nDuplicados: {df.duplicated().sum()}")
```

## Criterio 4: Calidad de los datos
El dataset no presenta valores nulos ni registros duplicados.

```python
print("--- CRITERIO 5: ESPECIFICIDAD DE RESTAURACIÓN ")
print(f"Restaurantes distintos: {df['restaurant_name'].nunique()}")
print(f"Tipos de cocina: {df['cuisine_type'].nunique()}")
print(f"\nTipos de cocina disponibles:")
print(df['cuisine_type'].value_counts())
```

## Criterio 5: Especificidad de restauración
El dataset incluye 178 restaurantes distintos de 14 tipos de 
cocina diferentes en Nueva York. No representa un único 
establecimiento ni está orientado específicamente a pizzerías 
con delivery.

## Verificación de criterios de evaluación

- ❌ **Volumen:** 1.898 registros repartidos entre 178 restaurantes 
  y 14 tipos de cocina distintos.
- ❌ **Fecha y hora:** sin fecha ni hora, solo indica 
  si el pedido es de fin de semana o entre semana.
- ❌ **Rango temporal:** imposible determinarlo al no disponer 
  de fecha real.
- ✅ **Calidad datos:** sin valores nulos ni duplicados.
- ❌ **Especificidad:** 178 restaurantes de 14 tipos de cocina 
  distintos, no representa un único establecimiento ni es 
  específico de pizzerías.

## Conclusión: dataset descartado

El análisis confirma que este dataset no cumple los criterios 
mínimos de selección.

La ausencia de fecha y hora es el criterio determinante para 
el descarte, ya que hace imposible construir una serie temporal 
y por tanto entrenar cualquier modelo predictivo.

**Decisión: dataset descartado.**
