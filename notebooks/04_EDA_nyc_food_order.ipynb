# EDA Dataset: NYC Restaurant Food Order & Delivery
## Análisis previo al descarte

Este notebook recoge el análisis exploratorio del dataset NYC Restaurant 
Food Order & Delivery como parte del proceso de selección del dataset 
para el TFG de predicción de demanda en pizzerías con delivery.

```python
import pandas as pd

# Carga del dataset
df = pd.read_csv('../data/kaggle_NYC_restaurants_data/food_order.csv')

print(f"Registros: {df.shape[0]}")
print(f"Columnas: {df.shape[1]}")
print(f"\nColumnas disponibles:")
print(df.columns.tolist())
print(f"\nPrimeras filas:")
print(df.head())
```

```python
# Análisis de la variable temporal
print(f"Valores únicos en day_of_the_week: {df['day_of_the_week'].unique()}")
print(f"\n¿Tiene columna de fecha? {'date' in df.columns}")
print(f"¿Tiene columna de hora? {'time' in df.columns or 'hour' in df.columns}")
print(f"\nDistribución por día:")
print(df['day_of_the_week'].value_counts())
```

## Conclusión: descarte del dataset

El análisis exploratorio confirma que este dataset no es adecuado 
para el objetivo del proyecto por los siguientes motivos:

- **Sin variable temporal real:** no dispone de fecha ni hora de 
  los pedidos. La única información temporal es si el pedido se 
  realizó en fin de semana o entre semana, lo que imposibilita 
  construir una serie temporal.
- **Granularidad insuficiente:** la variable `day_of_the_week` 
  solo tiene dos valores posibles, Weekend y Weekday, sin ningún 
  detalle adicional sobre el día concreto ni la franja horaria.
- **Volumen limitado:** solo 1.898 registros de múltiples 
  restaurantes de distintos tipos de cocina en Nueva York, sin 
  representar un único establecimiento.
- **Orientación incorrecta:** el dataset está diseñado para 
  analizar tiempos de preparación y entrega, no para predecir 
  volumen de demanda a lo largo del tiempo.

**Decisión: dataset descartado.**
