# EDA Dataset: Enhanced Pizza Sales 2024-2025
## Análisis previo al descarte

Este notebook recoge el análisis exploratorio del dataset Enhanced Pizza 
Sales 2024-2025 como parte del proceso de selección del dataset para el 
TFG de predicción de demanda en pizzerías con delivery.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carga del dataset
df = pd.read_excel('../data/kaggle_enhanced_pizza/Enhanced_pizza_sell_data_2024-25.xlsx')

print(f"Registros: {df.shape[0]}")
print(f"Columnas: {df.shape[1]}")
print(f"\nColumnas disponibles:")
print(df.columns.tolist())
```

## Análisis de la estructura

```python
# Primeras filas
print(df.head())
print(f"\nRango temporal:")
print(df['Order Time'].min())
print(df['Order Time'].max())

# Pedidos por día
df['Order Time'] = pd.to_datetime(df['Order Time'])
df['date'] = df['Order Time'].dt.date
pedidos_dia = df.groupby('date').size()
print(f"\nMedia de pedidos por día: {pedidos_dia.mean():.1f}")
print(f"Máximo pedidos en un día: {pedidos_dia.max()}")
print(f"Mínimo pedidos en un día: {pedidos_dia.min()}")
```

## Conclusión: descarte del dataset

El análisis exploratorio revela que este dataset no es adecuado para 
el objetivo del proyecto por los siguientes motivos:

- **Volumen insuficiente:** solo 1.004 registros repartidos entre 
  múltiples restaurantes y ciudades de Estados Unidos.
- **Densidad temporal inviable:** la media es de apenas 1,2 pedidos 
  por día, lo que hace imposible construir una serie temporal con 
  suficiente granularidad para entrenar un modelo predictivo.
- **Orientación incorrecta:** el dataset está diseñado para analizar 
  tiempos y eficiencia de entrega, no para predecir volumen de demanda.
- **Falta de coherencia de negocio:** los datos corresponden a 
  múltiples cadenas de pizzerías en distintas ciudades, sin 
  representar un único establecimiento.

**Decisión: dataset descartado.**
